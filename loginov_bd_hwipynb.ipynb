# Нормализация данных

* Таблица изначально находится в 1нф, т.к. атрибуты в таблице простые, все сохраняемые данные на пересечении столбцов и строк — содержать лишь скалярные значения.


* Чтобы привести таблицу к 2нф мы вывели атрибуты продукта в отдельную таблицу product из таблицу transaction. Больше никаких транзитивных зависимостей нет, postcode не является уникальным идентификатором, поэтому таблица теперь находится в 3нф.

# Создание таблиц

Судя по данным, поля в таблице могут быть пустыми, создадим таблицы следующей командой (product_id не указываем как primary key, т.к. судя по изначальным данным он может записывается отдельно и быть одинаковым для разных товаров).

```
CREATE TABLE customer (
    customer_id INTEGER PRIMARY KEY,
    first_name VARCHAR,
    last_name VARCHAR,
    gender VARCHAR,
    DOB DATE,
    job_title VARCHAR,
    job_industry_category VARCHAR,
    wealth_segment VARCHAR,
    deceased_indicator VARCHAR,
    owns_car VARCHAR,
    address VARCHAR,
    postcode INTEGER,
    state VARCHAR,
    country VARCHAR,
    property_valuation INTEGER
);

CREATE TABLE product (
    product_id INTEGER,
    brand VARCHAR,
    product_line VARCHAR,
    product_class VARCHAR,
    product_size VARCHAR,
    list_price FLOAT,
    standard_cost FLOAT
);

CREATE TABLE transaction (
    transaction_id INTEGER PRIMARY KEY,
    product_id INTEGER,
    customer_id INTEGER,
    transaction_date TIMESTAMP,
    online_order BOOLEAN,
    order_status VARCHAR,
    FOREIGN KEY (product_id) REFERENCES product (product_id),
    FOREIGN KEY (customer_id) REFERENCES customer (customer_id)
);
```
